# Kamień milowy 2(feature engineering and baseline)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA, TruncatedSVD
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.pipeline import Pipeline
```

```python
data = pd.read_csv("data.csv")
X = data.drop(["subject", "Activity", "ActivityName"], axis=1)
y = data["Activity"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.head()
```

```python
def drop_correlated(X, threshold=0.97):
    corr_df = X.corr().abs()
    columns_to_leave = np.ones((corr_df.shape[0], 1), dtype=bool)
    for i in range(corr_df.shape[0]):
        if columns_to_leave[i]:
            for j in range(i+1, corr_df.shape[0]):
                if corr_df.iloc[i, j] >= threshold:
                    columns_to_leave[j] = False
    columns_to_leave =  columns_to_leave.reshape(1, corr_df.shape[0])[0]
    return X.loc[:, columns_to_leave]
```

```python
X_train = drop_correlated(X_train)
X_train.head()
```

## Feature engineering 

```python
pca = PCA()
pca.fit(X_train)
explained_variance = pca.explained_variance_ratio_
n = 100

plt.rcParams["figure.figsize"] = [16,9]
plt.plot(np.cumsum(explained_variance[0:n]))
plt.bar(np.arange(n), explained_variance[0:n])
plt.xlabel("number of components")
plt.ylabel("cumulative explained variance");
plt.title("PCA")
plt.show()
```

```python
def reshape(components):
    res = np.zeros((2, components.shape[0]))
    for i in range(len(components)):
        res[0][i] = components[i][0]
        res[1][i] = components[i][1]
    return res
```

```python
pca = PCA(n_components = 2)
pca.fit(X_train)
pca_comp = pca.transform(X_train)
pca_comp = reshape(pca_comp)

plt.rcParams["figure.figsize"] = [20,15]
plt.scatter(pca_comp[0], pca_comp[1])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA")
plt.show()
# widać 2 klastry
```

```python
pipeline = Pipeline(
            [("pca", PCA(n_components = 2)),
            ("kmeans", KMeans(n_clusters = 3))])
pipeline.fit(X_train)

pca_data = pipeline["pca"].transform(X_train)
pca_data = reshape(pca_data)
plt.rcParams["figure.figsize"] = [20,15]
plt.scatter(pca_data[0], pca_data[1], c=pipeline["kmeans"].labels_)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA")
plt.show()
```

```python
pca = PCA(n_components = 50)
pca_comp = pca.fit_transform(X_train)
tsne = TSNE(n_components = 2, random_state=42)
tsne_comp = tsne.fit_transform(pca_comp)
tsne_comp = reshape(tsne_comp)

plt.rcParams["figure.figsize"] = [20,15]
plt.scatter(tsne_comp[0], tsne_comp[1])
plt.xlabel("comp1")
plt.ylabel("comp2")
plt.title("tSNE")
plt.show()
```

```python
plt.rcParams["figure.figsize"] = [20,15]
plt.scatter(tsne_comp[0], tsne_comp[1],  c=y_train)
plt.xlabel("comp1")
plt.ylabel("comp2")
plt.title("tSNE")
plt.show()
```
